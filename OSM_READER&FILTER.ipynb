```python
place_name = "Delft"
# graph = ox.graph_from_place(place_name)
# fig, ax = ox.plot_graph(graph)
# plt.tight_layout()

leisure = ox.geometries.geometries_from_place(place_name,  tags = {"natural":True})
# leisure = leisure[['leisure','name', 'sport', 'geometry']]
leisure = leisure[leisure['geometry'].type == 'Polygon']

crs = fiona.crs.from_epsg(2136)
geo = gpd.GeoDataFrame(leisure, geometry = 'geometry', crs=crs)

geo_m = geo.to_crs("EPSG:3395")
geo_m['Area'] = geo_m.area
geo_m['Perimeter'] = geo_m.length
# np.unique(leisure.leisure)

leisure
```
