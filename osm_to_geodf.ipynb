```python
import osmnx as ox
import geopandas as gpd
import fiona.crs
import numpy as np
import pandas as pd
import warnings
```

```python
warnings.filterwarnings('ignore')

place_name = "Δήμος Θεσσαλονίκης" #"Περιφέρεια Κεντρικής Μακεδονίας"
crs = fiona.crs.from_epsg(2100)
key = 'leisure'
value = 'park|garden|playground|dog_park|recreation_ground'

def osm_to_geodf(place_name, crs, key, value):
    data = ox.geometries.geometries_from_place(place_name,  tags = {key:True})
    geo_data = gpd.GeoDataFrame(data, geometry = 'geometry', crs=crs)
    geo_data = geo_data[geo_data[key].str.contains(value, na=False)]
    return geo_data

osm_to_geodf(place_name, crs, key, value)
```
